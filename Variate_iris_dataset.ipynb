```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
##Univariate Analysis
df_setosa = df.loc[df['variety']=='Setosa']
```

```python
df_virginica = df.loc[df['variety']=='Virginica']
df_versicolor = df.loc[df['variety']=='Versicolor']
```

```python
plt.plot(df_setosa['sepal.length'],np.zeros_like(df_setosa['sepal.length']),'o')
plt.plot(df_virginica['sepal.length'],np.zeros_like(df_virginica['sepal.length']),'o')
plt.plot(df_versicolor['sepal.length'],np.zeros_like(df_versicolor['sepal.length']),'o')
plt.xlabel('Petal length')
plt.show()

#here np.zeros is used to make all y axis 0. as here univariate analysis is used so y axis is not required.
```

```python
#Bivariate Analysis:
#seaborn will be used here.
sns.FacetGrid(df,hue="variety",size=5).map(plt.scatter,"sepal.length","sepal.width").add_legend();
plt.show()
#hue is used to differentiate the species.
#Facetgrid is used to plot the graph.
```

```python
sns.FacetGrid(df,hue="variety",size=5).map(plt.scatter,"petal.length","petal.width").add_legend();
plt.show()
```

```python
#Multivariate Analysis
#Pairplot is used to plot the graph.
sns.pairplot(df,hue="variety",size=3) 
```
