```python
import numpy as np
import pandas as pd
df = pd.read_csv('cancer_dataset.csv') #read the dataset
df.head()
```

```python
 # Independent and dependent features
X = df.iloc[:,2:] # independent features(it starts from 2nd column i.e. radius_mean
                   # till the last )
                   #iloc is used to select the rows and columns by number.
                   # Syntax:iloc[rows,columns]
                   #[:,2:] means all the rows and columns from 2nd column till the last
y= df.iloc[:,1] # dependent features (diagnosis)
                #[:,1] means all the rows and 1st column
           
```

```python
X.head()
```

```python
y.head()
```

```python
#the last column is redundant as all the values are NaN. So we can drop it.
X=X.dropna(axis=1) #dropna() is used to drop the rows or columns with NAN values
                   #axis=1 means columns
                   #axis=0 means rows
X.head()
```

```python
y.value_counts() #count the number of values in each class
```

```python
#HOLD OUT VALIDATION 
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier #import the decision tree classifier
from sklearn.model_selection import train_test_split #import the train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size = 0.30,random_state=0) 
#split the dataset into train and test
#test_size = 0.30 means 30% of the dataset is used for testing and 70% for training
#random_state=0 means the dataset is split randomly
#random_state=20 means the dataset is split randomly but the same split is used every time
model=DecisionTreeClassifier() #create the object of the DecisionTreeClassifier
model.fit(X_train,y_train)     #train the model using the training dataset
result=model.score(X_test,y_test) #test the model using the testing dataset
print(result)
```

```python
# K FOLD CROSS VALIDATION
from sklearn.model_selection import KFold  #import KFold
model=DecisionTreeClassifier()  #create the object of the DecisionTreeClassifier
kfold_validation=KFold(10) #10 splits are taken
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score 
#import cross_val_score
#cross_val_score is used to evaluate a score by cross-validation.
# It is used in K-fold cross validation but not in hold out validation 
results = cross_val_score(model,X,y,cv=kfold_validation) #cv is used to specify the cross validation splitting strategy
print(results)
print(np.mean(results))
#max accuracy-0.96491228
#min accuracy-0.89473684
```

```python
from sklearn.model_selection import StratifiedKFold #import StratifiedKFold
#mostly used for imbalanced dataset.(having 2 classes)
skfold=StratifiedKFold(n_splits=5)
model=DecisionTreeClassifier()
scores = cross_val_score(model,X,y,cv=skfold)
print(scores)
print(np.mean(scores))
```

```python
#LEAVE ONE OUT CROSS VALIDATION
#provides train/test indices to split data in train/test sets.
#Each sample is used once as a test set (singleton) while the remaining samples form the training set.
from sklearn.model_selection import LeaveOneOut
model=DecisionTreeClassifier()
leave_validation = LeaveOneOut()
results = cross_val_score(model,X,y,cv=leave_validation)
```
